```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

from google.colab import files
uploaded = files.upload()
```

```python
# 데이터 불러오기
df = pd.read_excel('202501.xlsx')
```

```python
# 시간 데이터로 변환 (Pump-Begin 기준)
df['Pump-Begin'] = pd.to_datetime(df['Pump-Begin'])
```

```python
# PM2.5 농도 데이터 전처리
df = df[df['Conc(ug/m3)'] >= 0] # 음수값 제거
df = df.dropna(subset=['Conc(ug/m3)']) # 결측치 제거
```

```python
# 날짜 컬럼 생성 (datetime을 date로 변환)
df['Date'] = df['Pump-Begin'].dt.date
```

```python
# 날짜를 11일 단위로 분할
unique_dates = sorted(df['Date'].unique())
chunks = [unique_dates[i:i+11] for i in range(0, len(unique_dates), 11)]
```

```python
# 11일 단위별 그래프 생성
for idx, date_group in enumerate(chunks, start=1):
    temp_df = df[df['Date'].isin(date_group)]  # 해당 구간 데이터만 추출

    # 막대그래프 그리기
    plt.figure(figsize=(12, 6))
    plt.bar(temp_df['Pump-Begin'], temp_df['Conc(ug/m3)'], width=0.03)

    # 그래프 기본 설정
    plt.title(f'PM2.5 Concentration ({date_group[0]} ~ {date_group[-1]})')
    plt.xlabel('Time')
    plt.ylabel('PM2.5 Concentration (ug/m³)')

    # x축 시간 설정
    plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=1))
    plt.xticks(rotation=45)

    # 그래프 출력
    plt.tight_layout()
    plt.show()
```
